```python
import easyocr
import cv2
import numpy as np
```

```python


reader = easyocr.Reader(['ru'], gpu=False)
img = cv2.imread('test.png')
result = reader.readtext(img)
for (bbox, text, prob) in result:
    print(f"{text} ({prob:.2f})")

```
